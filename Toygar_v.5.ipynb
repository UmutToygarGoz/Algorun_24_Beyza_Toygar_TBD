```python
'''
Yapılan submissionlar ve sonuç olarak alınan hatalar şu şekilde

second => 0.49731
(second + fourth) / 2 => 0.49827
(2*second + fourth + seventh) => 0.49547

bu nedenle 

(fourth + seventh) / 2 => daha iyi sonuç vermesini bekleriz. Ninth submissionu böyle yapacağım


'''


fourth_submission_df = pd.read_csv('fourth_submission.csv', delimiter="|")
seventh_submission_df = pd.read_csv('seventh_submission.csv', delimiter="|")

ninth_submission_df = sample_submission_df.copy()
ninth_submission_df["prediction"] = (fourth_submission_df["prediction"] + seventh_submission_df["prediction"]) / 2

# Making negative numbers in prediction equal to 0
ninth_submission_df.loc[ninth_submission_df["prediction"] < 0, "prediction"] = 0

ninth_submission_df.fillna(0,inplace=True)

ninth_submission_df.to_csv('ninth_submission.csv', index=False, sep="|")
```

```python
# Tenth submission, fourth submission'ın oranlarını eighth submission'ın toplamına uyarlayarak oluşturulacak.

fourth_submission_df = pd.read_csv('fourth_submission.csv', delimiter="|")
eighth_submission_df = pd.read_csv('eighth_submission.csv', delimiter="|")

fourth_submission_df['week_starting_date'] = pd.to_datetime(fourth_submission_df['week_starting_date'])
eighth_submission_df['week_starting_date'] = pd.to_datetime(eighth_submission_df['week_starting_date'])


# Calculate ratios from fourth submission for each product per week
fourth_weekly_totals = fourth_submission_df.groupby('week_starting_date')['prediction'].sum()
fourth_submission_df['sales_ratio'] = fourth_submission_df['prediction'] / fourth_submission_df.groupby('week_starting_date')['prediction'].transform('sum')

# Get weekly totals from eighth submission
eighth_weekly_totals = eighth_submission_df.groupby('week_starting_date')['prediction'].sum()

# Create new predictions using ratios from fourth and totals from eighth
tenth_submission_df = fourth_submission_df.copy()
tenth_submission_df['prediction'] = tenth_submission_df.apply(
    lambda row: row['sales_ratio'] * eighth_weekly_totals[row['week_starting_date']], 
    axis=1
)

# Drop the ratio column and ensure correct format
tenth_submission_df = tenth_submission_df.drop('sales_ratio', axis=1)

tenth_submission_df.fillna(0,inplace=True)

tenth_submission_df.loc[tenth_submission_df["prediction"] < 0, "prediction"] = 0


# Save to CSV if needed
tenth_submission_df.to_csv('tenth_submission.csv', sep='|', index=False)
```

```python
# Eleventh submission as an ensemble of (second + fourth + seventh)/4


import pandas as pd
second_submission_df = pd.read_csv('second_submission.csv',delimiter="|")
fourth_submission_df = pd.read_csv('fourth_submission.csv',delimiter="|")
seventh_submission_df = pd.read_csv('seventh_submission.csv', delimiter="|")

eleventh_submission_df = second_submission_df.copy()
eleventh_submission_df["prediction"] = (second_submission_df["prediction"] + fourth_submission_df["prediction"] + seventh_submission_df["prediction"]) / 3

# Making negative numbers in prediction equal to 0
eleventh_submission_df.loc[eleventh_submission_df["prediction"] < 0, "prediction"] = 0

eleventh_submission_df.fillna(0,inplace=True)

eleventh_submission_df.to_csv('eleventh_submission.csv', index=False, sep="|")
```
